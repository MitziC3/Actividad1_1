```python
#Reemplazo
df['CVE_CLPV'] = df['CVE_CLPV'].replace('M', '0') #Comando para remplazar un dato
#Conversión de tipo de dato de columna de tipo Object a Float
df['CVE_CLPV'] = df['CVE_CLPV'].astype(float)
```
